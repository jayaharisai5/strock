```python
#Model Validation

def get_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
X = np.array(X)
y = np.array(y)
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

folds = StratifiedKFold(n_splits=3)

RandomForestClassifier_score = []
KNeighborsClassifier_score = []
DecisionTreeClassifier_score = []

for train_index, test_index in folds.split(X,y):
    X_train, X_test, y_train, y_test = X[train_index],X[test_index],y[train_index],y[test_index]
    
    RandomForestClassifier_score.append(get_score(RandomForestClassifier(), X_train, X_test, y_train, y_test))
    KNeighborsClassifier_score.append(get_score(KNeighborsClassifier(), X_train, X_test, y_train, y_test))
    DecisionTreeClassifier_score.append(get_score(DecisionTreeClassifier(), X_train, X_test, y_train, y_test))
    
    
print("RandomForestClassifier_score: " + str(RandomForestClassifier_score))
print("KNeighborsClassifier_score: " + str(KNeighborsClassifier_score))
print("DecisionTreeClassifier_score: " + str(DecisionTreeClassifier_score))
```
